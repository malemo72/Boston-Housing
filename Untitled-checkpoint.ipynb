```python
Boston House Prices Dataset was collected in 1978 and has 506(final 6 rows) entries with 14 attributes or features for homes from various suburbs in Boston

Attributes given in the dataset i order
Crim  per capita crime rate by town(crime rate occuring in that particular town)
Zn    Proportion of residential land zoned for lots over 25,000 sq.ft
Indus Proportion of non-retail business acres per town
CHAS  Charles River dummy variable( = 1 if tract bounds river(if it has river it has one if otherwise zero); it is 0 if otherwise)
Nox   nitrix oxides concentration (parts per 10 million)
RM    average number of rooms per dwelling(no of rooms in a house)
Age   proportion o owner-occupied units build prior to 1940(age of the house)
Dis   weighted distances to five Boston employment centres(to govt offices etc to )
Rad   index of accesibility to radial highways ( how easy to access highway from the city)
Tax   full value property tax rate per $10,000(how much tax to pay for that particular property)
PTRATIO pupil-teacher ratio by town(how manu teachers and students are there)
B     1000(kinda racists or something in blacks comparidon)(Bk- 0.63)^2 where Bk is the proportion of blacksby town()
Lstat  % lower status of the population
Medv   Median value of owner-occupied homes in $1000's
```

# 5th step: Min-max Normalization

```python
cols = ['crim', 'zn', 'tax', 'b']
for col in cols:
        #find minimum and maximum of that column
        minimum = min(df[col])
        maximum = max(df[col])
        df[col] = (df[col] - minimum) / (maximum - minimum)
```

```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()

for col, value in df.items():
    sns.distplot(value, ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
#standardization
from sklearn import preprocessing
scalar = preprocessing.StandardScaler()

# fit our data
scaled_cols = scalar.fit_transform(df[cols])
scaled_cols = pd.DataFrame(scaled_cols, columns=cols)
scaled_cols.head()
```

```python
for col in cols:
    df[col] = scaled_cols[col]
```
